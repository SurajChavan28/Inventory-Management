```python
import json
from datetime import datetime

with open('Records.json', 'r') as fd:
    data = json.load(fd)


def add_products():
    count = 0
    while True:
        inp = input("Enter (Serial No.,Product Name,Price,Quantity): ").split(",")
        if len(inp) == 4:
            if inp[0] not in data.keys():
                data[inp[0]] = {'Name': inp[1], 'Price': int(inp[2]), 'Quantity': int(inp[3])}
                count += 1
            else:
                print("Choose a different serial countber")
                continue
            if input('Again Do you want to add the item: ').lower() == 'y':
                continue
            else:
                break
        else:
            print("Improper Input Format")
            continue
    return count


def sell_products():
    sold_items = []
    while True:
        inp = input("Enter (Serial No., Quantity) that you want to buy").split(',')
        if len(inp) == 2:
            if inp[0] in data.keys():
                data[inp[0]]['Quantity'] = int(data[inp[0]]['Quantity']) - int(inp[1])
                sold_items.append({"srno": inp[0], "count": int(inp[1])})
            else:
                print('Invalid Data')
                continue
            if input("Do u want to order anything (y/n): ").lower() == "y":
                continue
            else:
                break

        else:
            print('Improper Input Format')
    return sold_items


fd = open('Records.json', 'w')
if input("Want to add data(y/n) ").lower() == "y":
    count_prod = add_products()
    print(f"{count_prod} Products added...")
    fd.write(json.dumps(data))
    fd.close()


# to buy the product
for i in data.keys():
    print(f"{i}\t{data[i]['Name']}\t\t{data[i]['Price']}\t\t{data[i]['Quantity']}")
sold = sell_products()
t_amount = 0
sales_dc = dict()
for item in sold:
    t_amount += int(data[item["srno"]]["Price"]) * int(item["count"])

if 30000 <= t_amount < 40000:
    q = 5
    dis = t_amount * 0.05
    f_total = t_amount - dis
if t_amount >= 40000:
    q = 10
    dis = t_amount * 0.05
    f_total = t_amount - dis
else:
    f_total = t_amount
    q = 0

#billing
print('*'*50)  
print('Date',datetime.now())
print('*'*50)
for item in sold:
    print(f"Your product is: {data[item['srno']]['Name']}\nPrice of product: {data[item['srno']]['Price']}\nQuantity: {item['count']}")
print(f'You got the Discount of {q}%\nTotal Amount is: {f_total}\n', '*'*50,sep="")
fd = open('Records.json', 'w')
a = json.dumps(data)
fd.write(a)
fd.close()

sales = {f"{datetime.now()}": {'items': sold, 'Price': f_total, 'Quantity': [i['count'] for i in sold]}} 
e = json.dumps(sales)
fd = open('Sales.json', 'a')
fd.write(e)
fd.close()
```
